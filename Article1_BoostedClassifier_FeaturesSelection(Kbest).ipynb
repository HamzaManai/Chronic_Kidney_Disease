#                           **Chronic Kidney Disease Prediction**




## Introduction 
CKD is a common condition that refers to a long-term loss of kidney function, affecting millions of people globally. Over two million people worldwide receive dialysis, kidney transplant, or another type of treatment to stay alive, reaching only 10% of cases. Most of the time, CKD is detected during kidney failure, leading to extreme measures like kidney transplant, and death if they do not result. CKD has five stages of kidney damage, from very mild damage in stage 1 to complete kidney failure in stage 5, as can be seen in Figure .

image.png

In order to prevent people with this disease from reaching the ESKD,
techniques for predicting the disease in the early stages become an advantage and can save many human lives. In
the healthcare field, machine learning has been widely used to predict and classify clinical variables using the data
stored in the Hospital Information Systems (HIS)

## Methodology
This study used the CRISP-DM methodology, which defines a non-rigid sequence of five phases that allows the construction and implementation of a DM model to be used in a real environment, helping to support business decisions. The CRISP-DM lifecycle is composed by the following stages: Business Understanding, Data Understanding,
Data Preparation, Modeling, Evaluation and Deployment

image.png

## 1 -  Business Understanding


Early prediction and proper treatments can possibly stop or slow the progression of CKD to end-stage, where
dialysis or kidney transplantation are the only way to save patients life [1]. It is becoming increasingly important to
use this type of predictions since it is based on real patients information, allowing the creation of prediction models
that assist health professionals in their decisions. In this context, this study aims to predict cases of early stage CKD
based on clinical data through the use of DM techniques.

## 2 - Data Understanding
To carry out this study, it was used a dataset related to the detection of CKD, collected in a hospital in India over
a period of 2 months. The dataset has information of 400 patients and 24 important features in identifying risk
factors for CKD and improving the diagnosis of cases of patients with CKD. The dataset contains personal data, data
from laboratory tests on blood and urine and data on the patients clinical history. The attributes of the dataset are

# Naive bayes: base method 

```python
from sklearn.naive_bayes import BernoulliNB 
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score
import math
clf = BernoulliNB()  
acc=[]
prec=[]
rec=[]
fm=[]
rmse=[]
skf=StratifiedKFold(n_splits=5,random_state=None)
skf.get_n_splits(X,y)
for train_index,test_index in skf.split(X,y) :
  X1_train,X1_test = X.iloc[train_index],X.iloc[test_index]
  y1_train,y1_test=y.iloc[train_index],y.iloc[test_index]
  clf.fit(X1_train,y1_train.values.ravel())
  prediction=clf.predict(X1_test)
  accuracy=accuracy_score(y1_test,prediction)
  acc.append(accuracy)
  precision = precision_score(y1_test,prediction)
  prec.append(precision)
  recall = recall_score(y1_test,prediction)
  rec.append(recall)
  Fmeas = f1_score(y1_test,prediction)
  fm.append(Fmeas)
  acc1= np.mean(acc)
  prec1=np.mean(prec)
  rec1=np.mean(rec)
  fm1=np.mean(fm)
  MSE = np.square(np.subtract(y1_test,prediction)).mean()   
  rsme = math.sqrt(MSE)

  rmse.append(rsme)
  rmse1=np.mean(rmse)
 
print('Cross Validation accuracy: ' , (acc1))
print('Cross Validation precision: ' , (prec1))
print('Cross Validation recall: ' , (rec1))
print('Cross Validation F_measure: ' , (fm1))
print('rmse score: ' , (rmse1))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
#  Confusion Matrix
cm1 = confusion_matrix(y1_test,prediction)

disp = ConfusionMatrixDisplay(confusion_matrix=cm1)
disp.plot()
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc 

fig, (ax2) = plt.subplots(figsize = (6,4))
        #roc-curve
fpr, tpr, thresholds_roc = roc_curve(y1_test,prediction)
roc_auc = auc(fpr,tpr)
ax2.plot(fpr,tpr, label = " AUROC = {:0.2f}".format(roc_auc))
ax2.plot([0,1], [0,1], 'r', linestyle = "--", lw = 2)
ax2.set_xlabel("False Positive Rate", fontsize = 14)
ax2.set_ylabel("True Positive Rate", fontsize = 14)
ax2.set_title("ROC Curve", fontsize = 18)
ax2.legend(loc = 'best')
plt.title('ROC curve for Naive Bayess ')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
        #find default threshold
close_default = np.argmin(np.abs(thresholds_roc - 0.5))
ax2.plot(fpr[close_default], tpr[close_default], 'o', markersize = 8)
plt.tight_layout()
```
